```python
import pandas as pd 
import numpy as np
```

```python
boook_df=pd.read_csv('book.csv',encoding='ISO-8859-1')
```

```python
boook_df
```

```python
boook_df.reset_index(drop=True,inplace=True)
```

```python
boook_df.rename(columns={'User.ID': 'userID','Book.Title': 'booktitle','Book.Rating': 'bookrating'},inplace=True)
```

```python
boook_df
```

```python
user_book_df=boook_df.pivot_table(index='userID',
                           columns='booktitle',
                           values='bookrating').reset_index(drop=True)
user_book_df
```

```python
user_book_df.index=boook_df.userID.unique()
```

```python
user_book_df
```

```python
user_book_df.fillna(0, inplace=True)
user_book_df
```

```python
#calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
#pairwise distances give us dissimilarity, so that is why we do 1-pairwise distances
similarity_user=1-pairwise_distances(user_book_df.values,metric='cosine')
```

```python
similarity_user
```

```python
similarity_user_df=pd.DataFrame(similarity_user)
```

```python
similarity_user_df.index=boook_df.userID.unique()
similarity_user_df.columns=boook_df.userID.unique()
```

```python
similarity_user_df.iloc[0:10,0:10]
```

```python
np.fill_diagonal(similarity_user, 0)
similarity_user_df.iloc[0:5,0:5]
```

```python
similarity_user_df.idxmax(axis=1)[0:10]
```

# Item based Collaborative Filtering


```python
user_book_df.rename(columns={'Classical Mythology':'ClassicalMythology'},inplace=True)
user_book_df
```

```python
classical_mythology=user_book_df['ClassicalMythology']
classical_mythology
```

```python
similar_to_cm=user_book_df.corrwith(classical_mythology)
```

```python
similar_to_cm.head(10)
```

```python
corr_cm=pd.DataFrame(similar_to_cm,columns=['correlation'])
corr_cm.head()
```

```python
corr_cm[corr_cm['correlation']>0].sort_values(by='correlation', ascending=False).head(5)
```
